```python
from src.datasets import TSPLibDataset
from src.solvers import TSPRandomSolver, TSPAttentionModelSolver
import pandas as pd
from tqdm.notebook import tqdm
```

```python
dataset = TSPLibDataset()
random_solver = TSPRandomSolver()
attention_solver = TSPAttentionModelSolver.load_from_checkpoint('checkpoints/epoch=0-step=16.ckpt')
```

```python
results = {
    'name': [],
    'tsplib_cost': [],
    'random_cost': [],
    'attention_cost': [],
}

for instance, solution in tqdm(dataset):
    results['name'].append(instance.name)
    tsplib_cost = None
    if solution is not None:
        tsplib_cost = instance.total_cost(solution)
    results['tsplib_cost'].append(tsplib_cost)
    random_cost = instance.total_cost(random_solver.solve(instance))
    results['random_cost'].append(random_cost)
    attention_cost = instance.total_cost(attention_solver.solve(instance))
    results['attention_cost'].append(attention_cost)
```

```python
pd.DataFrame(results)
```
